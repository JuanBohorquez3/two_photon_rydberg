```python
%matplotlib notebook
from qutip import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
from arc import *
```

```python
def projector(Hamiltonian, state_index):
    pj = np.zeros(Hamiltonian.shape,dtype=int)
    pj[state_index,state_index] = 1
    return Qobj(pj)
```

```python
Omega1 = 10j
Omega2 = 2j
Delta1 = 2
Delta2 = 6
H_q = Qobj(np.array([[0,Omega1/2,0],[Omega1.conjugate()/2,Delta1, Omega2/2],[0,Omega2.conjugate()/2, Delta2]],dtype=complex))

psi_0 = basis(3,0)
t_list = np.linspace(0,10,1000)

gamma = 0.7
dec_op = np.sqrt(gamma)*Qobj([[0,gamma,gamma],[0,0,0],[0,0,0]])

results = mesolve(H_q,psi_0,t_list,[dec_op],[])
```

```python
fig,ax = plt.subplots(1,1)

for i in range(psi_0.shape[0]):
    ax.plot(t_list,expect(projector(H_q,i),results.states),label=f"$P_{i}$")
ax.legend()
ax.set_ylabel("Population")
ax.set_xlabel("Time")
fig.show()
```

```python
fig,ax = plt.subplots(1,1)

for i in range(psi_0.shape[0]):
    ax.plot(t_list,expect(projector(H_q,i),results.states),label=f"$P_{i}$")
ax.legend()
ax.set_ylabel("Population")
ax.set_xlabel("Time")
fig.show()
```
