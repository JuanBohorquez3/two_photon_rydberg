```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from arc import *
from time import time
from scipy.optimize import curve_fit
from typing import List, Dict

from basics import *
from utility import *
from floquet_hamiltonian import *

pi=np.pi
tau=2*pi
```

```python
t_level = RydStateFS(49,3,5/2)
# t_level = RydStateFS(48,3,5/2)

max_det = tau*50e9
dl = 1
n_max=0

comp = {
    "max_det": max_det,
    "dl": dl,
    "n_max": n_max
}

levels, basis = build_basis(t_level, max_det, single_side=False, dl=dl)
```

```python
Edc_list = np.linspace(0,20,100)
Eac = 0
ellipticity = 0
theta = 0
field_omega = 4000

H0 = build_H0(basis, t_level, offset_zeeman=1e3,j_split=10e6)

energies, eigenstates =floquet_loop(
    basis,
    H0,
    Eac,
    ellipticity,
    field_omega,
    Edc=0,
    theta=theta,
    n_max=0,
    varied=("Edc", Edc_list)
)
```

```python
fig,axar = plt.subplots(1,2, figsize=(10,6))

for i, state in enumerate(basis):
    if [state.n, state.l, state.j] != [t_level.n, t_level.l, t_level.j+1]:
        continue
    axar[0].plot(Edc_list, energies[i,:]*1e-6/tau, label = state.ket())
    axar[1].plot(Edc_list, np.abs(eigenstates[i,i,:])**2, label = state.ket())
axar[0].set_title(f"Energies of {t_level.ket()}")
axar[1].set_title(f"State Probability of {t_level.ket()}")
axar[1].set_yscale("log")
axar[1].legend()
fig.show()
```

```python

thrsh = 5e-2
for i, state in enumerate(basis):
    if [state.n, state.l, state.j] != [t_level.n, t_level.l, t_level.j]:
        continue
    fig,axar = plt.subplots(1,2, figsize=(10,6))
    for j, statep in enumerate(basis):
        if any(eigenstates[i,j,:] > thrsh):
            axar[0].plot(Edc_list, energies[j,:]*1e-6/tau, label = statep.ket())
            axar[1].plot(Edc_list, np.abs(eigenstates[i,j,:])**2, label = statep.ket())
    fig.suptitle(state.ket())
    axar[0].set_title(f"Energies")
    axar[1].set_title(f"State Probabilities")
    axar[1].set_yscale("log")
    axar[1].set_ylim(thrsh*0.8,1.1)
    axar[1].legend()
    axar[1].axhline(1, ls=":")
    fig.show()
```

```python
fig,axar = plt.subplots(int(2*t_level.j+1),2, figsize=(12,6*int(2*t_level.j+1)))
for i, state in enumerate(basis):
    if [state.n, state.l, state.j] != [t_level.n, t_level.l, t_level.j]:
        continue
    ax_ind = int(t_level.j+state["mj"])
    axar[ax_ind,0].plot(Edc_list, (np.abs(eigenstates[i,:,:])**2).sum(0), label = state.ket())
    for j, sp in enumerate(basis):
        if any(eigenstates[i,j,:] > 1e-16):
            axar[ax_ind,1].plot(Edc_list, (np.abs(eigenstates[i,j,:])**2), label = sp.ket())
    axar[ax_ind,0].set_title(f"Total population")
    axar[ax_ind,1].set_title(f"Population")
    axar[ax_ind,1].set_yscale("log")
    axar[ax_ind,0].legend()
    #axar[ax_ind,1].legend()
    axar[ax_ind,1].axhline(1, ls=":")
    fig.tight_layout()
    fig.show()
```
